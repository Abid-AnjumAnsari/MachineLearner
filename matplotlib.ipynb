## OO METHOD

The previous plot was not ideal , we were using pyplot method directly 
using panda data frame.

#### Which one should we use (pyplot vs matplotlib) oo method ?

* When Plotting something quickly okay to use pyplot method. when we used all of this previously.
* When Plotting something advance use the OO method.

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease['age']>50]
over_50
```

```python
over_50.shape,len(over_50)
```

```python
over_50.head()
```

```python
##Pyplot method aka directly using yplot via panda df
## As this is somewhat advance so we can also use OO method
over_50.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
## OO method mixed with pyplot
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);

#ax.set_xlim([45,100])## set the limit from 45 to 100
```

```python
## OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])

#Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",xlabel="age",ylabel="cholestrol");
#Add a legend

## did not worked --- ax.legend(*scatter.legend_elements(),title="Target") will have to check with advanced matplotlib
ax.legend(*scatter.legend_elements(),title="Target");
#ax.legend()
#it will unpack all the legend elements from scatter plot

#add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="dotted");
```

```python
import matplotlib
print('matplotlib: {}'.format(matplotlib.__version__))
```

```python
over_50.head()
```

```python
#subplots of chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))

#Add data to ax0
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])

#Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",xlabel="age",ylabel="Cholestrol")

#Add a legend to ax0
#ax0.legend(*scatter.legend_elements(),title="Target")
ax0.legend();

#Add a mean line

ax0.axhline(over_50["chol"].mean(),linestyle="dashed");

#now for age vs thalach

scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])

ax1.set(title="Heart Disease and Cholestrol Levels",xlabel="age",ylabel="Max Heart rate")

#ax1.legend(*scatter.legend_elements(),title="target")
ax1.legend()

ax1.axhline(over_50["thalach"].mean(),linestyle="dotted");
```

```python
#since they have different x axis seperately each telling age we can use 
#plt.subplot(sharex=True)
```

```python
#subplots of chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

#Add data to ax0
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])

#Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",ylabel="Cholestrol")

#Add a legend to ax0
#ax0.legend(*scatter.legend_elements(),title="Target")
ax0.legend();

#Add a mean line

ax0.axhline(over_50["chol"].mean(),linestyle="dashed");

#now for age vs thalach

scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])

ax1.set(title="Heart Disease and Cholestrol Levels",xlabel="age",ylabel="Max Heart rate")

#ax1.legend(*scatter.legend_elements(),title="target")
ax1.legend()

ax1.axhline(over_50["thalach"].mean(),linestyle="dotted");


##Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

```python
plt.style.use("seaborn-whitegrid")

## OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")#it will change the color

#Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",xlabel="age",ylabel="cholestrol");
#Add a legend

## did not worked --- ax.legend(*scatter.legend_elements(),title="Target") will have to check with advanced matplotlib
#ax.legend(*scatter.legend_elements(),title="Target");
ax.legend()
#it will unpack all the legend elements from scatter plot

#add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="dotted");
```

```python
#Customising the y and x axis
#subplots of chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

#Add data to ax0
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")

#Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",ylabel="Cholestrol")

ax0.set_xlim([50,80])#change the x axis


#Add a legend to ax0
#ax0.legend(*scatter.legend_elements(),title="Target")
ax0.legend();

#Add a mean line

ax0.axhline(over_50["chol"].mean(),linestyle="dashed");

#now for age vs thalach

scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="winter")

ax1.set(title="Heart Disease and Cholestrol Levels",xlabel="age",ylabel="Max Heart rate")

ax1.set_xlim([50,80])#not necessary as both graph have same axis
ax1.set_ylim([60,200])#change the y axis


#ax1.legend(*scatter.legend_elements(),title="target")
ax1.legend()

ax1.axhline(over_50["thalach"].mean(),linestyle="dotted");


##Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
